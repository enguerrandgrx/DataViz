```python
df['Latitude'] = df['Latitude'].fillna(181)
df['Longitude'] = df['Longitude'].fillna(181)
```

```python
geolocator = Nominatim()
```

```python
type(geolocator.geocode(df['City'][0]))
```

```python
for i in range(len(df)):
    if(df['Latitude'][i] == 181 or df['Longitude'][i] == 181):
        location = geolocator.geocode(df['City'][i])
        if(type(location) == geopy.location.Location):
            df.loc[i, 'Latitude'] = location.latitude
            df.loc[i, 'Longitude'] = location.longitude
```

```python
df[['Longitude', 'Latitude']]
```

```python
df['text'] = df['Date'].dt.strftime('%B %-d'
                          ) + ', ' + df['Fatalities'].astype(str) + ' Fatalities'

data = [dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df['Longitude'],
        lat = df['Latitude'],
        text = df['text'],
        mode = 'markers',
        marker = dict( 
            size = df[df.Longitude < 0]['Fatalities'] ** 0.5 * 5,
            opacity = 0.75,
            color = 'rgb(215, 0, 0)')
        )]

layout = dict(
         title = 'Shooting Fatalities by Latitude/Longitude in United States <br>'
                 '<sub>Hover to View Date and Fatalitiess</sub>',
         geo = dict(
             scope = 'usa',
             projection = dict(type = 'albers usa'),
             showland = True,
             landcolor = 'rgb(250, 250, 250)',
             subunitwidth = 1,
             subunitcolor = 'rgb(217, 217, 217)',
             countrywidth = 1,
             countrycolor = 'rgb(217, 217, 217)',
             showlakes = True,
             lakecolor = 'rgb(255, 255, 255)')
         )

figure = dict(data = data, layout = layout)
plotly.plotly.iplot(figure)
```
